## 2. サーバーに接続

`HOST` と `PLAYER_NAME` を自分の設定に変えてね

```python
from py2hackCraft2.modules import Player, Volume, LocationFactory

# ===== ここを変更してね =====
HOST = "localhost"        # サーバーのアドレス
PORT = 25570              # ポート番号
PLAYER_NAME = "your_name" # 自分のプレイヤー名
# ==========================

player = Player(PLAYER_NAME)
player.login(HOST, PORT)
print(f"{PLAYER_NAME} でログインしました！")
```

## 3. プログラミング開始！

下のセルにコードを書いて実行してみよう

```python
# エンティティを取得
entity = player.get_entity("entity_name")

# 前に10ブロック進む
entity.forward(10)
```
